# Importing necessary Libraries

```python
import pandas as pd
import re
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,classification_report,confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.models import Word2Vec
```

# Importing dataset

```python
column_names_train=['Title','Genre','Description']
train_data = pd.read_csv("Data/train_data.txt",sep=':::',names=column_names_train,engine='python')
train_data.head()
```

```python
column_names_test=['Title','Description']
test_data = pd.read_csv("Data/test_data.txt",sep=':::',names=column_names_test,engine='python')
test_data.head()
```

```python
combined_data = pd.concat([train_data,test_data],ignore_index=True)
data = combined_data
data.tail()
```

# Data preprocessing and Features Extraction

```python
data= data.dropna()
split_desc = data["Description"].apply(lambda x : x.lower().split())
word2vec_model = Word2Vec(split_desc,vector_size= 100, window =5,min_count = 1,workers = 4)
```

```python
def get_sentence_embedding(sentence):
    words = re.findall(r'\w+', sentence.lower())
    vectors = [word2vec_model.wv[word] for word in words if word in word2vec_model.wv]
    return sum(vectors) / len(vectors) if vectors else [0] * 100
```

```python
data["embeddings"] = data["Description"].apply(get_sentence_embedding)
X = pd.DataFrame(data["embeddings"].to_list())
```

```python
X
```

```python
label_encoder = LabelEncoder()
data.loc[:, "genre_encoded"] = label_encoder.fit_transform(data["Genre"])
y = data["genre_encoded"]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
genre_counts = data["Genre"].value_counts()
plt.bar(genre_counts.index,genre_counts,color = 'red',)
plt.xlabel("Genre")
plt.ylabel("Count")
plt.title("Genre Distribution")
plt.xticks(rotation=90)
plt.show()
```

 # MODEL RANDOM FOREST CLASSIFIER

```python
X_train, X_test, y_train,y_test = train_test_split(X, y, test_size=0.2) 
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train,y_train)
prediction =model.predict(X_test)
precision = precision_score(y_test, prediction,average='weighted',zero_division=1.0)
recall = recall_score(y_test, predictions,average='weighted',zero_division=1.0)
accuracy = accuracy_score(y_test,prediction)
```

```python
print(f"Precisions: {precision}")
print(f"Recall: {recall}")
print(f"Accuracy: {accuracy}")
```

# Print the classification report


```python
print(classification_report(y_test, predictions))
```

# Print the confusion matrix

```python
confusion_mat = confusion_matrix(y_test, predictions)
print("Confusion Matrix:")
print(confusion_mat)
```

# Genre Prediction

```python
def predict_genre (model, label_encoder,new_descriptions):
    new_embeddings = [get_sentence_embedding(desc) for desc in new_descriptions]
    new_X = pd.DataFrame(new_embeddings)
    new_predictions = model.predict(new_X)
    predicted_genres = label_encoder.inverse_transform(new_predictions)
    
    return predicted_genres

new_descriptions = ["A group of friends embark on an adventurous journey.","In a dystopian future, a hero rises to save the world.","A heartwarming story of family and friendship."]
predicted_genres = predict_genre(model, label_encoder,new_descriptions)

for desc , genre in zip(new_descriptions, predicted_genres):
    print(f"Description: {desc}")
    print(f"Predicted Genre: {genre}")
    print()
```
